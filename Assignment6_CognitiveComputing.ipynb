```python
import numpy as np
import matplotlib.pyplot as plt

N = int(input("Enter a value: "))  # Changed variable name
x = np.linspace(-8, 8, 100)  # Slightly changed range
y1 = N * (x**2)
y2 = N * np.sin(x)

plt.plot(x, y1, label="Quadratic: y = Nx^2", linestyle="--", color="g")  # Modified style
plt.plot(x, y2, label="Sine: y = Nsin(x)", linestyle="-.", color="m")  # Modified style

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph of Quadratic and Sine Functions")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

N = pd.DataFrame({'Subjects': ['Physics', 'Chemistry', 'Biology', 'English', 'Music'],
                  'Scores': [8, 95, 82, 38, 57]})

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subjects', y='Scores', hue='Subjects', data=N, palette='coolwarm', width=0.6, legend=False)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                ha='center', va='bottom', fontsize=10, color='black')

plt.title('Student Scores by Subject')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(987654321)

random_data = np.random.randn(50)
cumsum_values = np.cumsum(random_data)

fig, axes = plt.subplots(2, 2, figsize=(8, 6))

axes[0, 0].plot(cumsum_values, marker='s', color='blue', linestyle='--')
axes[0, 0].set_title('Cumulative Sum')
axes[0, 0].grid(True)

axes[0, 1].scatter(range(50), random_data, color='green', marker='x')
axes[0, 1].set_title('Random Scatter')
axes[0, 1].grid(True)

axes[1, 0].axis('off')
axes[1, 1].axis('off')

plt.tight_layout(pad=3)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import files

uploaded_files = files.upload()

sales_df = pd.read_csv("company_sales_data.csv")
sales_df.head()

plt.plot(sales_df['month_number'], sales_df['total_profit'], marker='d', color='r', linestyle='-')
plt.title("Total Profit Over Months")
plt.xlabel("Month"), plt.ylabel("Total Profit")
plt.grid()
plt.show()

for col in sales_df.columns[1:-1]:
    plt.plot(sales_df['month_number'], sales_df[col], marker='d', linestyle='-', label=col)

plt.title("Product Sales Over Time")
plt.xlabel("Month"), plt.ylabel("Sales")
plt.legend(), plt.grid()
plt.show()

sales_df.set_index('month_number').plot(kind='bar', figsize=(8, 6), alpha=0.8, colormap='viridis')
plt.title("Sales Data Overview")
plt.grid()
plt.show()
```
